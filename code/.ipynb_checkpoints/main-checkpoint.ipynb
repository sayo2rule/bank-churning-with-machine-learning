### Step 4: Data Processing

#### Categorical Column Encoding

```python
# Import the needed modules; scikit-learn to apply encoding methods
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
# For the geography column (3 unique values), use one-hot encoding
# For the gender colum(2 unique), use label encoding

# Create the encoding objects.
onehot_encoder = OneHotEncoder()
label_encoder = LabelEncoder()
```

```python
# Reshape the geography column.
geog_reshaped = np.array(df["Geography"]).reshape(-1, 1)

# Apply the appropriate encoding method to each of the non-numeric column
geog_values = onehot_encoder.fit_transform(geog_reshaped)
gend_values = label_encoder.fit_transform(df["Gender"])
```

```python
#   Obtain the resulting data to ensure that it is in the desired format and that all categories are correctly encoded
print("Geography column after encoding:\n", geog_values.toarray()[25:35])
print("\nGender column after encoding:\n", list(gend_values)[25:35])
```

```python
# Create DataFrames from the encoded values
geog_encoded_df = pd.DataFrame(geog_values.toarray(), columns=(list(df["Geography"].unique())))
gend_encoded_df = pd.DataFrame(gend_values, columns=['Gender_Encoded'])

# Concatenate the encoded DataFrames with the original DataFrame at a specific column
insert_column_index = -1  # Define the index where you want to insert the new columns
df2 = pd.concat([df.iloc[:, :insert_column_index], geog_encoded_df, gend_encoded_df, df.iloc[:, insert_column_index:]], axis=1)


# Now, 'df2' contains the original columns along with the encoded values
# Explore the data to understand its structure
df2.head(10)
```

```python
# Save the processed data as "Churn_Modelling_Processed" to a csv file
df2.to_csv("../data/processed/Churn_Modelling_Processed.csv")

# Reload the processed dataset into a Pandas dataframe using the read_csv function
df2 = pd.read_csv("../data/processed/Churn_Modelling_Processed.csv")
```

#### Feature Selection

```python
# Select only the numeric columns for correlation calculation
numeric_columns = df2.select_dtypes(include=['number'])

# Calculate the correlation matrix with the "Exited" column
correl_matrix = numeric_columns.corr()

# Sort and visualize feature correlations with the target variable
correl_with_target = correl_matrix['Exited'].sort_values(ascending=False)

# Display the features with their correlation matrices
print("Features and their matrices:")
print(correl_with_target.head(20))

# Visualize the features with their correlation matrices
features = correl_with_target.head(20)

# Create a horizontal bar chart
plt.figure(figsize=(6, 4))
plt.barh(features.index, features.values)
plt.xlabel('Correlation')
plt.ylabel('Features')
plt.title("Features with 'Exited'")
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest correlation at the top
plt.show()
```

### Model Building and Evaluation

#### Splitting of Dataset

```python
# Import the needed modules for spliting of dataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Split the data into features (X) and the target variable (y)
y = df2['Exited']

# Remove non-numeric columns and 'Exited' from the dataset
non_numeric = ['Surname', 'Geography', 'Gender']
X = df2.drop(columns=['Exited'] + non_numeric)  # Exited is the target variable

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Feature scaling (standardization)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

#### Regression Models

```python
# Import the needed modules for the regression
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Define and train three regression models
models = [
    ('Linear Regression', LinearRegression()),
    ('Decision Tree Regression', DecisionTreeRegressor()),
    ('Random Forest Regression', RandomForestRegressor())
]

for model_name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calculate evaluation metrics
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    # Print or display results
    print(f'{model_name}:')
    print(f'Mean Absolute Error: {mae}')
    print(f'Mean Squared Error: {mse}')
    print(f'R-squared: {r2}')
    print('\n')
```

#### Classification Models

```python
# Import the needed modules
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Define and train three classification models
models = [
    ('Logistic Regression', LogisticRegression()),
    ('Decision Tree Classifier', DecisionTreeClassifier()),
    ('Random Forest Classifier', RandomForestClassifier())
]

for model_name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calculate evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    
    # Print or display results
    print(f'{model_name}:')
    print(f'Accuracy: {accuracy}')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'F1-score: {f1}')
    print(f'ROC-AUC: {roc_auc}')
    print('\n')
```
